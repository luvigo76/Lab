Al aplicarle un potencial al cátodo este expulsa electrones con una velocidad $v$ determinada. Debido a la disposición de las bobinas se crea un campo magnético, perpendicular a la velocidad del electrón, que lo desvía de su trayectoria describiendo así una curva. La fuerza magnética a la que está sometida el electrón:
$$
F_m = F_c \Rightarrow qvB = ma_c \Rightarrow qvB = m\frac{v^2}{r}
$$



Finalmente, obtenemos una expresión que nos permite calcular la relación entre la carga y la masa del electrón mediante una regresión lineal:
$$
\frac{v}{r} = \frac{q}{m}B
$$

Para obtener el radio de la curvatura debemos utilizar geometría básica guiándonos por la siguiente imagen:

![Esquema](Esquemabueno.png)

El radio de la circunferencia descrita por la partícula será:
$$r=\frac{k^2+f^2}{\sqrt{2}(k-f)}(m)$$

Mientras que el campo magnético podemos calcularlo a partir de la fórmula utilizada para las bobinas de Helmholtz:
$$B=\left(\frac{4}{5}\right)^{\frac{3}{2}}\frac{\mu_0n}{R}I=4.2I\left(\frac{mT}{A}\right)$$

Podemos ver que para determinar el campo magnético, únicamente necesitamos conocer el número de vueltas y la intensidad que pasa por estas.

Por último, debemos obtener la velocidad de las partículas que como hemos dicho antes será para todas la misma. 

Gracias a las placas podemos crear un campo eléctrico de forma que la fuerza eléctrica se opone a la magnética. Cuando la trayectoria de las partículas vuelva a ser una línea recta entonces tendremos:
$$
F_E = F_M \Rightarrow qE = qvB \Rightarrow v = \frac{E}{B}
$$


Lo que nos permite determinar la velocidad de las partículas, el valor del campo eléctrico podemos obtenerlo cómo: $$E = \frac{V}{d}$$

 Conocida la velocidad, la curvatura y el campo magnético ya podremos determinar la relación carga-masa del electrón.

```python
import numpy as np
import matplotlib.pyplot as plt


I=np.array([0.7,0.75,0.79,0.84,0.88,0.94,1.08]) #Valores experimentales de la corriente en Amperios.

V=np.array([360,380,400,420,440,460,520])#Valores experimentales del voltaje en Voltios.

f=np.array([3.6e-2,3.4e-2,3.2e-2,3e-2,2.8e-2,2.6e-2,2e-2]) #Valores experimentales en metros para calcular el radio (ver imagen).

E=V/(8e-3) #Valores del campo eléctrico (V/m).

r=((80e-3)**2+f**2)/(np.sqrt(2)*(80e-3-f)) #Valores del radio (m).

B=4.2e-3*I #Valores del campo magnético (T). Este será el eje x.

vel=E/B #Valores de la velocidad (m/s).

Y=(vel)/r #Este será el eje y.
```

```python
# Realizar la regresión lineal
m, b = np.polyfit(B, Y, 1)

# Imprimir los coeficientes de la regresión
print(f"Pendiente (C/kg): {m}")
print(f"Intersección: {b}")


#Gráfica de los resultados y la regresión lineal
plt.plot(B, m*B + b, label='Ajuste lineal')
plt.scatter(B, Y, color='red', label='Datos')
plt.xlabel('Campo magnético (T)')
plt.ylabel('Velocidad/Radio (1/s)')
plt.legend()
plt.show()
```

En cuanto al resultado obtenido, podemos ver la relación carga-masa de los electrones sería $0.377 \times 10^{11} (C/kg)$, mientras que la relación real es $1.759 \times 10^{11} (C/kg)$. 

El error relativo es de $78\% $.
